# Beispiel mit pyfst

```python
import fst
t=fst.Acceptor()
```

pyfst ist ein Wrapper um OpenFST (eine Bibliothek, die in C++ geschrieben ist). Mit pyfst kann man endliche Automaten und endliche Transducer in Python konstruieren. Zusätzlich kann pyfst die entstandenen endlichen Automaten auch (im iPython Notebook) nett anzeigen.

Erst einmal setzen wir die Zustandsübergänge ein und markieren Zustand 1 als Endzustand (0 ist per Konvention der Anfangszustand).

```python
t.add_arc(0,0,'a')
t.add_arc(0,1,'b')
t.add_arc(1,1,'c')
t[1].final=True
```

Wenn wir einfach nur t ausgeben lassen, zeigt iPython den Transducer grafisch an.

```python
t
```

... und wir können zufällige Symbolfolgen aus dem Automaten generieren:

```python
t.uniform_generate()
```
